```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
app_data = pd.read_csv("Data/google-play-store-apps/googleplaystore.csv")
review_data = pd.read_csv("Data/google-play-store-apps/googleplaystore_user_reviews.csv")
```

```python
# Cleaning data to drop all cells with missing info
clean_app_data = app_data.dropna(how="any")
clean_app_data
```

```python
clean_app_data.count()
```

```python
categories = app_data["Category"].unique()
average_rating_by_category = app_data.groupby("Category")["Rating"].mean()
```

```python
fix, ax = plt.subplots(figsize=(10,8))
grid = sns.barplot(average_rating_by_category, categories, orient="h")
plt.show()
```

```python
# adjusted data to account for discrepancy in Art data
app_scrub = app_data[app_data["Category"] != "1.9"]["Category"].unique()
review_scrub = app_data[app_data["Category"] != "1.9"].groupby("Category")["Rating"].mean()
```

```python
fix, ax = plt.subplots(figsize=(12,12))
sns.barplot(app_scrub, review_scrub, orient="v")
plt.xticks(rotation=90)
plt.show()
```
